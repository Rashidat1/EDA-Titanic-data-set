```python
#Creating a line plot and scatter plot with Matplotlib.

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(url)

# Scatter plot
plt.scatter(df["age"], df["fare"], alpha=0.6)
plt.title("Age vs Fare (Scatter Plot)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Task : Line plot
# To make a line plot, first sort by age to make the line meaningful
df_sorted = df.sort_values("age")

plt.plot(df_sorted["age"], df_sorted["fare"], color="orange")
plt.title("Age vs Fare (Line Plot)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Task : Line plot using average value 
avg_fare_by_age = df.groupby("age")["fare"].mean()

plt.plot(avg_fare_by_age.index, avg_fare_by_age.values, color="green")
plt.title("Average Fare by Age")
plt.xlabel("Age")
plt.ylabel("Average Fare")
plt.show()
```
